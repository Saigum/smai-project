```python
import pandas as pd
import glob, os
from sklearn.metrics import accuracy_score, precision_score, recall_score, classification_report
from ultralytics import YOLO
model = YOLO('/scratch/saigum/runs/classify/yolov8n-kaggle-fullrun5/weights/best.pt')
results= model.predict(
    source='images_val/images_val',  # <-- only the val folder
    task='classify',
    conf=0.25,
    save=True,
    save_txt=True,
    save_conf=True,
    name='yolo11x-cls-kaggle-predict',
    device=0
)
print(results)
```

```python

# 1) load true labels and zeroâ€‘index them
trues = pd.read_csv('labels_val.csv', usecols=['Region_ID', 'filename'])
trues['true_class'] = trues['Region_ID']
trues = trues.drop(columns='Region_ID')
# strip extension if present so it matches the txt filenames
trues['filename'] = trues['filename'].str.replace(r'\.\w+$', '', regex=True)

# 2) parse all topâ€‘5 prediction files
records = []
label_files = glob.glob('/scratch/saigum/runs/classify/yolo11x-cls-kaggle-predict/labels/*.txt')
for file_path in label_files:
    fname = os.path.basename(file_path).replace('.txt','')
    with open(file_path) as f:
        for rank, line in enumerate(f.read().splitlines(), start=1):
            prob, cls = line.split()
            records.append({
                'filename':    fname,
                'rank':        rank,
                'class_id':    int(cls),
                'probability': float(prob)
            })

df = pd.DataFrame(records)

# 3) get topâ€‘1 predictions & merge with true labels
top1 = df[df['rank'] == 1].merge(trues, on='filename', how='left')
y_true = top1['true_class']
y_pred = top1['class_id']
# 4) compute metrics for topâ€‘1
accuracy  = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='macro', zero_division=0)
recall    = recall_score(y_true, y_pred, average='macro', zero_division=0)

print(f"Topâ€‘1 Accuracy:  {accuracy:.4f}")
print(f"Macroâ€‘Precision: {precision:.4f}")
print(f"Macroâ€‘Recall:    {recall:.4f}\n")

print("Classification Report (Topâ€‘1):")
print(classification_report(y_true, y_pred, zero_division=0))

top1.to_csv('yolo_11xtop1.csv', index=False)


```
